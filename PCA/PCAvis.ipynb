```python
from sklearn import datasets
import pandas as pd

# load dataset
iris = datasets.load_iris()
features = iris.data
target = iris.target

# generate a pandas dataframe
df = pd.DataFrame(features)
df.columns = iris.feature_names
```

```python
print(target)
print(iris.target_names)
```

```python
print(df.shape)
print(df.head(20))
```

```python
# print dataset descriptions and class distributions
print(df.describe())
```

```python
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt

# display scatter_matrix 
scatter_matrix(df, figsize=(10, 10))
plt.show()
```

```python
# Elbow method to determine optimal number of clusters
from sklearn.cluster import KMeans

# Empty x and y data lists
X = []
Y = []

for i in range(1, 31):
    # initialize and fit kernel models
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(df)
    
    # append number of clusters x data list
    X.append(i)
    
    # append average within-cluster sum of squares to y data list
    awcss = kmeans.inertia_ / df.shape[0]
    Y.append(awcss)
```

```python
import matplotlib.pyplot as plt

# Plot the X and Y data
plt.plot(X, Y, 'bo-')
plt.xlim((1, 30))
plt.xlabel('Number of Clusters')
plt.ylabel('Average within-cluster sum of squares')
plt.title('K-Means clustering Elbow Method')
plt.show()
```

```python
# In the above diagram, the elbow is 3 clusters. So we'll use 3 clusters in our PCA
from sklearn.decomposition import PCA
from sklearn import preprocessing
```

```python
# Perform PCA
pca = PCA(n_components = 2)
pc = pca.fit_transform(df)

# Print new dimensions
print(pc.shape)
print(pc[:10])
```

```python
# re-fit the kmeans model to the principal components with the appropriate number of clusters
kmeans = KMeans(n_clusters = 3)
kmeans.fit(pc)
```

```python
# Visualize high-dimensional clusters using principal components
import numpy as np
# set size for the mesh
h = 0.02

# generate mesh grid
x_min, x_max = pc[:, 0].min() - 1, pc[:, 0].max() + 1
y_min, y_max = pc[:, 1].min() - 1, pc[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# label each point in mesh using last trained model
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.figure(figsize = (12, 12))
plt.clf()
plt.imshow(Z, interpolation = 'nearest', extent = (xx.min(), xx.max(), yy.min(), yy.max()),
          cmap = plt.cm.tab20c,
          aspect = 'auto',
          origin = 'lower')

# Plot the principal components on the color plot
for i, point in enumerate(pc):
    if target[i] == 0:
        plt.plot(point[0], point[1], 'g.', markersize=10)
    if target[i] == 1:
        plt.plot(point[0], point[1], 'r.', markersize=10)
    if target[i] == 2:
        plt.plot(point[0], point[1], 'b.', markersize=10)

# Plot the cluster centroids
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], marker = 'x', s = 250, linewidth = 4, color = 'w', zorder=10)

# set plot title and axis limits
plt.title('K-Means clustering on PCA-reduced Iris Data set')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.xticks(())
plt.yticks(())

# display
plt.show()
```

```python
from sklearn import metrics

# K-means clustering on non-reduced data
kmeans1 = KMeans(n_clusters = 3)
kmeans1.fit(features)

# Kmeans clustering on reduced data
kmeans2 = KMeans(n_clusters = 3)
kmeans2.fit(pc)

# Print metrics from non-reduced data
print('Non-reduced data:')
print('Homogeniety: {}, Completeness: {}, V-measure: {}'.format(metrics.homogeneity_score(target, kmeans1.labels_), 
                                                                metrics.completeness_score(target, kmeans1.labels_), 
                                                                metrics.v_measure_score(target, kmeans1.labels_)))

# Print metrics for reduced data
print('Reduced data:')
print('Homogeniety: {}, Completeness: {}, V-measure: {}'.format(metrics.homogeneity_score(target, kmeans2.labels_), 
                                                                metrics.completeness_score(target, kmeans2.labels_), 
                                                                metrics.v_measure_score(target, kmeans2.labels_)))
```
